Przygotuj funkcję `anova()`, która będzie weryfikować hipotezę o równości wartości przeciętnych i wyświetlać wynik testu statystycznego zgodnie z wytycznymi zawartymi w treści zadania oraz w docstring funkcji.

Przykład komunikatu zawartego w tytule figury:</br>

Uwaga! Celem zadanie jest zapoznanie się ze schematem metody i samodzielne napisanie funkcji od podstaw. W związku z tym podczas rozwiązywania zadania nie należy korzystać z gotowych funkcji przeprowadzających analizę wariancji.

```python
def anova(data, alpha):
    """
    data: DataFrame
        Tablica zawierająca domyślny indeks i dowolnie nazwane kolumny (nazwy prób losowych) z wynikami prób losowych. Znajdujące się w tablicy wartości NaN nie są wynikami wchodzącymi w skład prób losowych,
        a jedynie uzupełniają puste miejsca w tablicy powstałe na skutek różnej liczebności poszczególnych prób.
    alpha: float
        Wartość określająca poziom istotności.
    
    Returns
    -------
    F: float
        Wyliczona na podstawie próby losowej wartość statystyki F.
    F_alpha: float
        Wartość statystyki F_alpha.
    H: int
        Wynik testu statystycznego, przyjmuje wartość:
        0 - gdy wynik testu istotności nie daje podstaw do odrzucenia H0 na rzecz H1 na poziomie istotności 1-alpha,
        1 - gdy następuje odrzucenie H0 na rzecz H1 na poziomie istotności 1-alpha.
    """
    # YOUR CODE HERE
    groups = [data[column].dropna().values for column in data.columns]


    all_data = np.concatenate(groups)
    overall_mean = np.mean(all_data)
    total_sum_squares = np.sum((all_data - overall_mean) ** 2)

    # Calculate the between-group sum of squares
    group_means = [np.mean(group) for group in groups]
    between_group_sum_squares = np.sum([len(group) * (mean - overall_mean) ** 2 for group, mean in zip(groups, group_means)])

    # Calculate the within-group sum of squares
    within_group_sum_squares = np.sum([(value - group_mean) ** 2 for group, group_mean in zip(groups, group_means) for value in group])

    # Calculate the degrees of freedom
    num_groups = len(groups)
    num_total = len(all_data)
    df_between = num_groups - 1
    df_within = num_total - num_groups

    # Calculate the mean squares
    mean_square_between = between_group_sum_squares / df_between
    mean_square_within = within_group_sum_squares / df_within

    # Calculate the F-statistic
    F = mean_square_between / mean_square_within

    # Calculate the critical F-value
    F_alpha = sp.stats.f.ppf(1 - alpha, df_between, df_within)

    # Determine the result of the hypothesis test
    H = 1 if F > F_alpha else 0

    # Generate the plot
    x = np.linspace(sp.stats.f.ppf(0.0001, df_between, df_within), sp.stats.f.ppf(0.9999, df_between, df_within), 100)
    y = sp.stats.f.pdf(x, df_between, df_within)

    plt.plot(x, y, 'b-', label='F-distribution')
    plt.fill_between(x, y, where=(x > F_alpha), color='red', alpha=0.3)
    plt.axhline(y=np.max(y) / 3, color='green', linestyle='--')
    plt.title('H0: Wartości przeciętne nie różnią się od siebie w sposób istotny statystycznie\n'
              'H1: Co najmniej dwie wartości przeciętne różnią się od siebie w sposób istotny statystycznie\n'
              f'F = {F:.4f}, F_alpha = {F_alpha:.4f}\n'
              f'{"" if F < F_alpha else "!"}F {"<" if F < F_alpha else ">="} F_alpha\n'
              f'{"Wynik testu istotności nie daje podstaw do odrzucenia H0 na rzecz H1" if F < F_alpha else "Odrzucenie H0 na rzecz H1"} '
              f'na poziomie istotności 1-alpha = {1-alpha:.2f}')
    plt.xlabel('F-value')
    plt.ylabel('Probability Density')
    plt.legend(loc='best')
    plt.show()

    return F, F_alpha, H
    #raise NotImplementedError()
```
