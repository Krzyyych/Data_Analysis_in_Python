Zwizualizuj funkcje gęstości prawdopodobieństwa następujących rozkładów teoretycznych:

```python
# YOUR CODE HERE
fig, ax = plt.subplots(2, 3, figsize = [15, 10])

colors = ['blue', 'red', 'green', 'purple', 'orange']
x = np.linspace(-10, 10, 500)

means = [0, 0, 0, -2, 2]
stds = [1, 0.75, 1.25, 1, 1]
for i in range(len(means)):
    ax[0][0].plot(x, sp.stats.norm.pdf(x, loc = means[i], scale = stds[i]), color = colors[i])
ax[0][0].legend(['N(0,1)', 'N(0, 0.75)', 'N(0, 1.25)', 'N(-2, 1)', 'N(2, 1)'], title='Distribution')
ax[0][0].set_xlim(-5, 5)
ax[0][0].set_ylim(0.0, 0.8)
ax[0][0].set_xlabel('X')
ax[0][0].set_ylabel('pdf(X)')
ax[0][0].set_title('Rozkłady normalne')

dfs = [1, 2, 5, 30]
ax[0][1].plot(x, sp.stats.norm.pdf(x, loc = means[0], scale = stds[0]), color = colors[0])
for i in range(len(dfs)):
    ax[0][1].plot(x, sp.stats.t.pdf(x, df = dfs[i]), colors[i+1])
ax[0][1].legend(['N(0,1)', 'T(1)', 'T(2)', 'T(5)', 'T(30)'], title='Distribution')
ax[0][1].set_xlim(-5, 5)
ax[0][1].set_ylim(0.0, 0.8)
ax[0][1].set_xlabel('X')
ax[0][1].set_ylabel('pdf(X)')
ax[0][1].set_title('Rozkłady normalny i t-Studenta')

dfs2 = [1, 2, 3, 5, 8]
for i in range(len(dfs2)):
    ax[0][2].plot(x, sp.stats.chi2.pdf(x, df = dfs2[i]), colors[i])
ax[0][2].legend(['chi2(1)', 'chi2(2)', 'chi2(3)', 'chi2(5)', 'chi2(8)'], title='Distribution') 
ax[0][2].set_xlim(0, 10)
ax[0][2].set_ylim(0.0, 0.8)
ax[0][2].set_xlabel('X')
ax[0][2].set_ylabel('pdf(X)')
ax[0][2].set_title('Rozkłady chi-kwadrat')

dfds = [1, 2, 5, 10, 20]
for i in range(len(dfds)):
    ax[1][0].plot(x, sp.stats.f.pdf(x, dfn = 10, dfd = dfds[i]), colors[i])
ax[1][0].legend(['F(10, 1)', 'F(10, 2)', 'F(10, 5)', 'F(10, 10)', 'F(10, 20)'], title='Distribution') 
ax[1][0].set_xlim(0, 5)
ax[1][0].set_ylim(0.0, 1.6)
ax[1][0].set_xlabel('X')
ax[1][0].set_ylabel('pdf(X)')
ax[1][0].set_title('Rozkłady F Snedecora stałe dfn = 10')

dfns = [1, 2, 5, 10, 20]
for i in range(len(dfns)):
    ax[1][1].plot(x, sp.stats.f.pdf(x, dfn = dfns[i], dfd = 10), colors[i])
ax[1][1].legend(['F(1, 10)', 'F(2, 10)', 'F(5, 10)', 'F(10, 10)', 'F(20, 10)'], title='Distribution') 
ax[1][1].set_xlim(0, 5)
ax[1][1].set_ylim(0.0, 1.6)
ax[1][1].set_xlabel('X')
ax[1][1].set_ylabel('pdf(X)')
ax[1][1].set_title('Rozkłady F Snedecora stałe dfd = 10')

dfs3 = [1, 2, 5, 25, 50]
for i in range(len(dfns)):
    ax[1][2].plot(x, sp.stats.f.pdf(x, dfn = dfs3[i], dfd = dfs3[i]), colors[i])
ax[1][2].legend(['F(1, 1)', 'F(2, 2)', 'F(5, 5)', 'F(25, 25)', 'F(50, 50)'], title='Distribution') 
ax[1][2].set_xlim(0, 5)
ax[1][2].set_ylim(0.0, 1.6)
ax[1][2].set_xlabel('X')
ax[1][2].set_ylabel('pdf(X)')
ax[1][2].set_title('Rozkłady F Snedecora stałe dfn = dfd')


fig.suptitle('Podstawowe rozkłady teoretyczne wykorzystywane w statystyce matematycznej', fontsize = 14)
#raise NotImplementedError()
```

#### a) Wyliczanie prawdopodobieństw na podstawie wartości

Oblicz i zapisz do zmiennych o podanych nazwach następujące wartości prawdopodobieństwa:
- `p1` - $P(X < -1.5)$, gdzie $X$ to zmienna losowa ciągła posiadająca rozkład t-Studenta o 10 stopniach swobody - $t(10)$,
- `p2` - $P(X > 2)$, gdzie $X$ to zmienna losowa ciągła posiadająca rozkład F Snedecora o 5 i 50 stopniach swobody - $F(5, 50)$,
- `p3` - $P(3 < X < 10)$, gdzie $X$ to zmienna losowa ciągła posiadająca rozkład chi-kwadrat o 4 stopniach swobody - $\chi^2(4)$,
- `p4` - $P(X < -2.75 | X > 1.5)$, gdzie $X$ to zmienna losowa ciągła posiadająca rozkład normalny o wartości oczewiwanej równej -0.5 i odchyleniu standardowym równym 1.5 - $N(-0.5, 1.5)$,

Każde z rozwiązań zilustruj wykresem zawierającym krzywą gęstości prawdopodobieństwa rozkładu z zaznaczonym kolorem fragmentem pola pod krzywą odpowiadający wyznaczonej wartości prawdopodobieństwa. Przygotuj figurę zgodnie z wytycznymi z nagłówka notatnika. Dodatkowo:
- ustaw zakres osi x od wartości zmiennej $X$, dla której dystrybuanta analizowanego rozkładu przyjmuje wartość 0.0001 do wartości zmiennej $X$, dla której dystrybuanta analizowanego rozkładu przyjmuje wartość 0.9999,
- w tytule wykresu zawrzyj uzupełniony wzór na prawdopodobieństwo uzupełniony o wyliczoną wartość prawdopodobieństwa podaną do 4 miejsca po przecinku (np. P(X < -1.5) = x.xxxx, gdzie x.xxxx to obliczona wartość).

```python
# p1
# YOUR CODE HERE
t = sp.stats.t(10)
p1 = t.cdf(-1.5)

x = np.linspace(t.ppf(0.0001), t.ppf(0.9999), 500)
fig, ax = plt.subplots(1, 1, figsize = [5, 5])
ax.plot(x, t.pdf(x), color = 'black')
ax.fill_between(x, t.pdf(x), color = 'green', where = x < -1.5)
ax.set_title(f'P(X < -1.5) = {p1:.4f}')
ax.set_xlabel('X')
ax.set_ylabel('pdf(X)')
#raise NotImplementedError()
```

```python
# p2
# YOUR CODE HERE
f = sp.stats.f(5, 50)
p2 = 1 - f.cdf(2) #f.sf(2)

x = np.linspace(f.ppf(0.0001), f.ppf(0.9999), 500)
fig, ax = plt.subplots(1, 1, figsize = [5, 5])
ax.plot(x, f.pdf(x), color = 'black')
ax.fill_between(x, f.pdf(x), color = 'green', where = x > 2)
ax.set_title(f'P(X > 2) = {p2:.4f}')
ax.set_xlabel('X')
ax.set_ylabel('pdf(X)')
#raise NotImplementedError()
```

```python
# p3
# YOUR CODE HERE
chi2 = sp.stats.chi2(4)
p3 = chi2.cdf(10) - chi2.cdf(3)

x = np.linspace(chi2.ppf(0.0001), chi2.ppf(0.9999), 500)
fig, ax = plt.subplots(1, 1, figsize = [5, 5])
ax.plot(x, chi2.pdf(x), color = 'black')
ax.fill_between(x, chi2.pdf(x), color = 'green', where = (x > 3) & (x < 10))
ax.set_title(f'P(3 < X < 10) = {p3:.4f}')
ax.set_xlabel('X')
ax.set_ylabel('pdf(X)')
#raise NotImplementedError()
```

```python
# p4
# YOUR CODE HERE
n = sp.stats.norm(-0.5, 1.5)
p4 = n.cdf(-2.75) +  (1 - n.cdf(1.5))

x = np.linspace(n.ppf(0.0001), n.ppf(0.9999), 500)
fig, ax = plt.subplots(1, 1, figsize = [5, 5])
ax.plot(x, n.pdf(x), color = 'black')
ax.fill_between(x, n.pdf(x), color = 'green', where = (x < -2.75) | (x > 1.5))
ax.set_title(f'P(X < -2.75 | X > 1.5) = {p4:.4f}')
ax.set_xlabel('X')
ax.set_ylabel('pdf(X)')
#raise NotImplementedError()
```

#### b) Wyliczanie wartości na podstawie prawdopodobieństw

Rozwiąż i zapisz do zmiennych o podanych nazwach rozwiązania następujących równań:
- `s1` - $P(X < s) = 0.25$, gdzie X to zmienna losowa ciągła posiadająca rozkład chi-kwadrat o 5 stopniach swobody - $\chi^2(5)$,
- `s2` - $P(X > s) = 0.15$, gdzie X to zmienna losowa ciągła posiadająca rozkład F Snedecora o 5 i 20 stopniach swobody - $F(5, 20)$,
- `s3` - $P(-1.25 < X < s) = 0.35$, gdzie x to zmienna losowa ciągła posiadająca rozkład normalny o wartości oczewiwanej równej -0.4 i odchyleniu standardowym równym 0.75 - $N(-0.4, 0.75)$,

Każde z rozwiązań zilustruj wykresem zawierającym krzywą gęstości prawdopodobieństwa rozkładu z zaznaczonym kolorem fragmentem pola pod krzywą odpowiadający wartości prawdopodobieństwa określonej przez rozwiązywane równanie. Przygotuj figurę zgodnie z wytycznymi z nagłówka notatnika. Dodatkowo:
- ustaw zakres osi x od wartości zmiennej $X$, dla której dystrybuanta analizowanego rozkładu przyjmuje wartość 0.0001 do wartości zmiennej $X$, dla której dystrybuanta analizowanego rozkładu przyjmuje wartość 0.9999,
- oznacz krawędź obszaru, na której znajduje się wyznaczona wartość $s$ linią w wyróżniającym się kolorze,
- w tytule wykresu zawrzyj uzupełniony wzór na prawdopodobieństwo uzupełniony o wyliczoną wartość zmiennej $X$ podaną do 4 miejsca po przecinku (np. P(X < s.ssss) = 0.25, gdzie s.ssss to obliczona wartość).

```python
# s1
# YOUR CODE HERE
chi2 = sp.stats.chi2(5)
s1 = chi2.ppf(0.25)

x = np.linspace(chi2.ppf(0.0001), chi2.ppf(0.9999), 500)
fig, ax = plt.subplots(1, 1, figsize = [5, 5])
ax.plot(x, chi2.pdf(x), color = 'black')
ax.fill_between(x, chi2.pdf(x), color = 'green', where = x < s1)
ax.set_title(f'P(X < {s1:.4f}) = 0.25')
ax.set_xlabel('X')
ax.set_ylabel('pdf(X)')
#raise NotImplementedError()
```

```python
# s2
# YOUR CODE HERE
f = sp.stats.f(5, 20)
s2 = f.ppf(0.85)

x = np.linspace(f.ppf(0.0001), f.ppf(0.9999), 500)
fig, ax = plt.subplots(1, 1, figsize = [5, 5])
ax.plot(x, f.pdf(x), color = 'black')
ax.fill_between(x, f.pdf(x), color = 'green', where = x > s2)
ax.set_title(f'P(X > {s2:.4f}) = 0.15')
ax.set_xlabel('X')
ax.set_ylabel('pdf(X)')
#raise NotImplementedError()
```

```python
# s3
# YOUR CODE HERE
n = sp.stats.norm(-0.4, 0.75)
s3 = n.ppf(0.35 + n.cdf(-1.25))

x = np.linspace(n.ppf(0.0001), n.ppf(0.9999), 500)
fig, ax = plt.subplots(1, 1, figsize = [5, 5])
ax.plot(x, n.pdf(x), color = 'black')
ax.fill_between(x, n.pdf(x), color = 'green', where = (x <= s3) & (x >= -1.25))
ax.set_title(f'P(-1.25 < X < {s3:.4f}) = 0.35')
ax.set_xlabel('X')
ax.set_ylabel('pdf(X)')
#raise NotImplementedError()
```
