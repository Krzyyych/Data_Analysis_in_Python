### Zadnie 2: Prezentacja danych w formie tabelarycznej [10 pkt]

Przygotuj trzy tablice szeregów rodzielczych przedziałowych częstości wskaźnika GDP growth (annual %) dla poszczególnych lat.

Pierwsza tablica (`recession_table_1`) powinna zawierać szeregi skonstruowane w obarciu o całość danych.

Indeks:
 - Year (wszystkie lata)
 
Kolumny:
 - Kolejne przedziały (-20, -15]	(-15, -10] ... (30, 35] zdefiniowane jako IntervalIndex

Druga tablica (`recession_table_2`) powinna zawierać szeregi skonstruowane indywidualnie dla każdego z rejonów geograficznych.

Indeks:
 - 1 poziomu: Region Name
 - 2 poziomu: Year (wszystkie lata dla każdego z rejonów)
 
Kolumny:
 - Kolejne przedziały (-20, -15]	(-15, -10] ... (30, 35] zdefiniowane jako IntervalIndex

Trzecia tablica (`recession_table_3`) powinna zawierać szeregi skonstruowane indywidualnie dla każdego stopnia zamożności.

Indeks:
 - 1 poziomu: Income Group
 - 2 poziomu: Year (wszystkie lata dla każdej z grup)
 
Kolumny:
 - Kolejne przedziały (-20, -15]	(-15, -10] ... (30, 35] zdefiniowane jako IntervalIndex
 
Wyświetl przygotowane tablice w taki sposób aby zawarte w tablicy liczby (za wyjątkiem lat i granic przedziałów w nagłówkach kolumn i wierszy) wyświetlały się z dokładnością do 2 miejsca po przecinku.

```python
# Całość danych
# YOUR CODE HERE
ii=pd.interval_range(start=-20, freq=5, end=35)
idx=["2006", "2007", "2008", "2009", "2010", "2011"]
recession_table_1=pd.DataFrame(index = idx, columns = ii)
for index, value in recession_data_1[idx].iteritems():
    recession_table_1.loc[index,:] = recession_data_1.loc[:, index].groupby(pd.cut(value, ii)).count()
    recession_table_1.loc[index] /= recession_table_1.loc[index].sum()
recession_table_1.style.format('{:.2f}')
#raise NotImplementedError()
```

```python
# Podział na rejony
# YOUR CODE HERE
ii=pd.interval_range(start=-20, freq=5, end=35)
idx=["2006", "2007", "2008", "2009", "2010", "2011"]
regions=recession_data_1['Region Name'].unique()

idx2 = pd.MultiIndex.from_product([regions, idx])
recession_table_2=pd.DataFrame(index = idx2, columns = ii)

for region in regions:
    for index, value in recession_data_1[idx].iteritems():
        recession_table_2.loc[region, index] = recession_data_1[recession_data_1['Region Name'] == region].loc[:,index].groupby(pd.cut(value, ii)).count()
        recession_table_2.loc[region, index] /= recession_table_2.loc[region, index].sum()
recession_table_2.style.format('{:.2f}')
#raise NotImplementedError()
```

```python
# Podział pod względem stopnia zamożności
# YOUR CODE HERE
ii=pd.interval_range(start=-20, freq=5, end=35)
idx=["2006", "2007", "2008", "2009", "2010", "2011"]
incomes=recession_data_1['Income Group'].unique()

idx2 = pd.MultiIndex.from_product([incomes, idx])
recession_table_3=pd.DataFrame(index = idx2, columns = ii)

for income in incomes:
    for index, value in recession_data_1[idx].iteritems():
        recession_table_3.loc[income, index] = recession_data_1[recession_data_1['Income Group'] == income].loc[:,index].groupby(pd.cut(value, ii)).count()
        recession_table_3.loc[income, index] /= recession_table_3.loc[income, index].sum()
recession_table_3.style.format('{:.2f}')
#raise NotImplementedError()
```

Wykres liniowy: jest czytelny i łatwy do interpretacji, w bardzo przystępny sposób można zauważyć trend dla danej zmiennej oraz zmianę wartości w czasie. Dla większej ilości regionów, a nie tylko kontynentów tworzy nam się plątanina linii co spowodowałoby nieczytelność.

Boxplot: Przedstawia dodatkowe informacje o danych jak średnia, kwartyle, wartości odstające i ich rozpiętość, cięższy do interpretacji przez osoby niezaznajomione ze statystyką.

Violinplot: Dla nas osobiście całkowicie nieczytelny, niezrozumiały, a do tego nawet brzydko wygląda. 

Swarmplot: Przedstawia dokładne wartości danych co przekłada się na dokładną analize, pozwala dosyć łatwo zidentyfikować wartości odstające. Nawet przy zbiorze danych jak ten, który analizujemy - komunikat o utracie niekotrych danych ze względu na za dużą ilość punktów, które analizujemy.

Stripplot: Reprezentuje dokładne wartości danych rozdzielone na słupki a'la klastry co ułatwia odczyt, łatwo wskazać wartości odstające. Przy dużym zagęszczeniu punktów (duża ilość danych) może zlewać się w "plamę".
